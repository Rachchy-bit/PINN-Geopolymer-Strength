```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# 1. DATA (With the Day 28 "Lab Error")
X_raw = np.array([1, 3, 7, 10, 14, 21, 28, 45, 60, 90]).reshape(-1, 1).astype(np.float32)
y_raw = np.array([10, 15, 25, 28, 30, 35, 20, 42, 45, 48]).reshape(-1, 1).astype(np.float32)
X_data, y_data = tf.constant(X_raw), tf.constant(y_raw)

# 2. RANDOM FOREST
rf = RandomForestRegressor(n_estimators=100, random_state=42).fit(X_raw, y_raw.ravel())

# 3. STANDARD ANN (Data Only - No Physics)
ann = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(32, activation='tanh'),
    tf.keras.layers.Dense(32, activation='tanh'),
    tf.keras.layers.Dense(1)
])
ann.compile(optimizer=tf.keras.optimizers.Adam(0.01), loss='mse')
ann.fit(X_raw, y_raw, epochs=1000, verbose=0)

# 4. PINN (Data + Physics Monotonicity)
pinn = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(1,)),
    tf.keras.layers.Dense(32, activation='tanh'),
    tf.keras.layers.Dense(32, activation='tanh'),
    tf.keras.layers.Dense(1)
])
opt = tf.keras.optimizers.Adam(0.01)

for i in range(1000):
    with tf.GradientTape(persistent=True) as tape:
        tape.watch(X_data)
        y_pred = pinn(X_data)
        data_loss = tf.reduce_mean(tf.square(y_data - y_pred))
        grads = tape.gradient(y_pred, X_data)
        physics_loss = tf.reduce_mean(tf.square(tf.minimum(grads, 0.0))) 
        total_loss = 1*data_loss + 4000.0 * physics_loss 
    g = tape.gradient(total_loss, pinn.trainable_variables)
    opt.apply_gradients(zip(g, pinn.trainable_variables))
    del tape

# 5. EVALUATION & VISUALIZATION
X_plot = np.linspace(1, 90, 150).reshape(-1, 1).astype(np.float32)
y_rf = rf.predict(X_plot)
y_ann = ann.predict(X_plot, verbose=0)
y_pinn = pinn.predict(X_plot, verbose=0)

# Calculate R2 on the noisy training data
r2_rf = r2_score(y_raw, rf.predict(X_raw))
r2_ann = r2_score(y_raw, ann.predict(X_raw, verbose=0))
r2_pinn = r2_score(y_raw, pinn.predict(X_raw, verbose=0))

plt.figure(figsize=(11, 7))
plt.scatter(X_raw, y_raw, color='black', marker='x', s=100, label='Lab Data (with Error)')
plt.plot(X_plot, y_rf, 'g--', label=f'Random Forest (R²={r2_rf:.3f})')
plt.plot(X_plot, y_ann, 'r:', label=f'Standard ANN (R²={r2_ann:.3f})', linewidth=2)
plt.plot(X_plot, y_pinn, 'b-', label=f'PINN (R²={r2_pinn:.3f})', linewidth=3)

plt.axvspan(24, 32, color='gray', alpha=0.1, label='Outlier Impact Zone')
plt.title('Comparison: RF vs ANN vs PINN on Noisy Engineering Data', fontsize=14)
plt.xlabel('Age (Days)')
plt.ylabel('Strength (MPa)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

print(f"--- R2 Comparison ---")
print(f"Random Forest: {r2_rf:.4f} (Overfit)")
print(f"Standard ANN:  {r2_ann:.4f} (Follows Error)")
print(f"PINN:          {r2_pinn:.4f} (Physically Correct)")
```

```python
# Test a value inside the "Error Zone" (e.g., Day 25)
test_day = np.array([[25.0]]).astype(np.float32)

val_ann = ann.predict(test_day, verbose=0)[0][0]
val_pinn = pinn.predict(test_day, verbose=0)[0][0]

print(f"\n--- Prediction at Day 25 (Between Day 21 and 28) ---")
print(f"Standard ANN: {val_ann:.2f} MPa (Likely dropping toward the error)")
print(f"PINN:         {val_pinn:.2f} MPa (Likely maintaining the upward trend)")
```
